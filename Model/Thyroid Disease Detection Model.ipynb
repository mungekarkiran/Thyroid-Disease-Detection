```python
names=['age','sex','on_thyroxine','query_on_thyroxine','on_antithyroid_medication',
       'sick','pregnant','thyroid_surgery','I131_treatment','query_hypothyroid','query_hyperthyroid',
      'lithium','goitre','tumor','hypopituitary','psych','TSH_measured','TSH','T3_measured','T3',
       'TT4_measured','TT4','T4U_measured','T4U','FTI_measured','FTI','TBG_measured','TBG','referral_source','classes']
```

```python
df = pd.read_table(r'data\allhypo.data', sep=',', names=names, na_values='?')
df.head()
```

```python
# split the class name
class_name = df['classes'].str.split('.', expand=True)
# class_name
df[['class_name','class_num']] = class_name
df.head()
```

```python
# drop classes and class_num columns
df.drop(['classes', 'class_num'], axis='columns', inplace=True)
df.head()
```
